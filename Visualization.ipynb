# **VISUALIZATION FOR DIABETES DATASET**

```
import pandas as pd
from google.colab import files
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```
db = pd.read_csv('diabetes.csv', index_col=0)
db.head()
```

## **Matplotlib**
# Scatter Plot

```
# create a figure and axis
fig, ax = plt.subplots()

# scatter the sepal_length against the sepal_width
ax.scatter(db['Age'], db['Glucose'])
# set a title and labels
ax.set_title('Diabetes Dataset')
ax.set_xlabel('age')
ax.set_ylabel('glucose')
```

# *Histogram*

```
# create figure and axis
fig, ax = plt.subplots()
# plot histogram
ax.hist(db['Age'])
# set title and labels
ax.set_title('Age groups')
ax.set_xlabel('Age')
ax.set_ylabel('Frequency')
```

# *Bar Chart*

```
# create a figure and axis 
fig, ax = plt.subplots() 
# count the occurrence of each class 
data = db['Age'].value_counts() 
# get x and y data 
points = data.index 
frequency = data.values 
# create bar chart 
ax.bar(points, frequency) 
# set title and labels 
ax.set_title('Bar chart') 
ax.set_xlabel('Age') 
ax.set_ylabel('Frequency')
```
